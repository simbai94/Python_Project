```python
class PiggyBank:

    def __init__(self, bal):
        self._balance = bal

    def open_account(self, amount):

        self._balance += amount

    def deposit(self, amount):

        self._balance += amount

    def withdraw(self, amount):

        if self._balance >= amount:
            self._balance -= amount
        else:
            print("Transaction denied! You have insufficient funds")

    def get_balance(self):
        return self._balance
```

```python
print("Welcome to Piggy Bank, If you don't have an account, kindly Open one and enjoy the convinience of banking with us")


def main():
    start_bal = 0
    savings = PiggyBank(start_bal)

    restart = 'Y'

    while restart not in ('n', 'NO', 'no', 'N'):
        print('Please Press 1 To Open Account\n')
        print('Please Press 2 To Make a Deposit \n')
        print('Please Press 3 To Make a Withdrawal \n')
        print('Please Press 4 To Check Balance \n')
        options = int(input("Select one option from the list above:  "))
        if options == 1:
            start_bal = float(input("Enter your initial deposit: "))
            savings.open_account(start_bal)
            print('Successfully opened Account. Your balance is $', format(savings.get_balance(), ',.2f'), sep='')
            
        elif options == 2:
            add = float(input("Enter your deposit amount: "))
            print(str(add) + ' has been credited to your account.')
            savings.deposit(add)
            print('Your account balance is $', format(savings.get_balance(), ',.2f'), sep='')
            
        elif options == 3:
            withdraw = float(input("Enter your withdrawal amount: "))
            print(str(withdraw) + ' has been withdrawn from your account.')
            savings.withdraw(withdraw)
            print('Your account balance is $', format(savings.get_balance(), ',.2f'), sep='')
            
        else:
            print('Your balance is $', format(savings.get_balance(), ',.2f'), sep='')
            
        restart = input ( 'Would You you like to go back? ')
        if restart in ('n', 'NO', 'no', 'N'):
            print('Thank You for banking with PiggyBank, your savings Partner')
            break
            
main()

```
